```python
raw_text=' '.join(df_whatsapp2['rawText'])
raw_text=raw_text.lower()
words=nltk.word_tokenize(raw_text)
```

```python
len(words)
```

```python
word_low=[]
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english')) 
from string import punctuation
for i in words:
    if i not in punctuation and i not in stop_words:
        word_low.append(i)
print('total words after puncations are removed',len(word_low))
custom_stopwords=['media','omitted','hmm','voice','call','deleted','message']
```

```python
custom_stopwords=['media','omitted','hmm','voice','call','deleted','message']
word_low=[x for x in word_low if x not in custom_stopwords ]
print('total words after puncations are removed',len(word_low))
```
